```python

""" 
Question 4 	 [40 points]
Plot and interpretation required
What is the survival ratio by age, sex, and class.
 
Note: All your visualisations will be marked also on correctness so please pay attention to the labels, axes, titles, etc. 
 
 
"""

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# reads  Titanic dataset
titanic = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv')

# calculates survival ratios
survival_ratios = titanic.groupby(['sex', 'class', 'age'])['survived'].mean().reset_index()

# defines customised colours using RGB values 
custom_palette = {'male':(250/255, 0, 128/255) ,  # pink RGB value
                  'female': (100/255, 0, 250/255)}# purple RGB value

# creates grouped bar plot with customised colours
plt.figure(figsize=(12, 6))
sns.barplot(data=survival_ratios, x='age', y='survived', hue='sex', ci=None,
            palette=custom_palette)
plt.title('Survival Ratio by Age, Sex, and Class')
plt.xlabel('Age')
plt.ylabel('Survival Ratio')
plt.legend(title='Sex', loc='upper right')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

#Interpretation:

"""
Looking at the graph there was no distinct upward or downward trend in survival ratios across different age groups. 
This becomes evident that  the age is a primary factor influencing these survival ratios.

A significant finding, is most female passengers tended to have higher survival rates compared to their male equivalents. 
This strong trend aligns with the well-known principle of "women and children first" .Throughout history, women and children
are prioritsed as they are seen as more weak and need protecting. Therefore in this situation, they were probably given 
priority towards the lifeboat during emergency, which directly correlates to the observed data.

Furthermore, the pattern is noticeable when considering class distinctions. Generally, first-class passengers presents 
higher survival ratios than those in the second or third class. This inconsistency is due to a higher priority  given to
those who coulld afford to pay for first-class passenger tickects during the evacuation process.

Interestingly, when we delve deeper into specific age groups, particularly among females, a consistent trend emerges where 
first-class passengers have notably higher survival rates than those in the second or third class.

In summary, the overarching trend supports the notion that during the evacuation, the prioritisation of women and children
led to higher survival rates, particularly among younger female passengers. It's clear that a combination of age and class
played a vital role in shaping the evacuation process, favouring certain demographic groups over others. This is evident in
significantly higher survival rates among females, especially in younger age groups.

The most significant evidence of these trends lies in the survival ratios of males and females within the younger age groups,
which provides a compelling demonstration of the prioritisation that occurred during the Titanic disaster.



"""

```
